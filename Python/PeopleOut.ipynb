```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
FileAddress = 'TimePeople.csv'
df=pd.read_csv(FileAddress)
df.head()
```

```python
#clean Time with no people change
df.drop(df[df.timestamp == '3/5/2022 7:00:00 AM'].index, inplace=True)
df.drop(df[df.timestamp == '3/5/2022 8:35:00 AM'].index, inplace=True)
df
```

```python
dfTime  = df.copy()
dfTime.drop(columns=['participantId'],axis=0, inplace=True)
dfTime=dfTime.drop_duplicates() #41 unique time
dfTime
```

```python
Time=dfTime['timestamp'].tolist()
Time
```

```python
Time.sort()
print(Time[0])
print(Time[1])
print(Time[-1])
```

```python
DiffAll = []
count = 0
for i in range(0,len(Time)-1):
    j = i+1
    #print(Time[i],Time[j])
    dfFirst = df[df['timestamp'] == Time[i]]
    First=dfFirst['participantId'].tolist()
    First.sort()
    #First
    dfSecond = df[df['timestamp'] == Time[j]]
    Second=dfSecond['participantId'].tolist()
    Second.sort()
    #Second
    Diff = list(set(First) - set(Second))
    if(Diff == []):
        print(Time[i])
    else:
        count+=1
    Diff.sort()
    DiffAll.append(Diff)
```
